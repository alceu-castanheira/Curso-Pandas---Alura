# Relatório de Atividades VI

## Criando Novas Variáveis

```python
# Importando Pandas, o banco de dados residencial filtrado da atividade anterior e verificando os dados
import pandas as pd

dados = pd.read_csv('dados/aluguel_residencial_2.csv', sep = ';')
dados.head(10)
```

```python
# Criando uma nova variável 'Valor bruto' = 'Valor' + 'Condominio' + 'IPTU'
dados['Valor Bruto'] = dados['Valor'] + dados['Condominio'] + dados['IPTU']
dados.head(10)
```

```python
# Criando uma nova variável 'Valor/m²' = 'Valor' / 'Area'. Round(2) arredonda o resultado para duas casas decimais
dados['Valor/m²'] = (dados['Valor'] / dados['Area']).round(2)
dados.head(10)
```

```python
# Criando uma nova variável 'Valor Bruto/m²' = 'Valor Bruto' / 'Area'. Round(2) arredonda o resultado para duas casas 
# decimais
dados['Valor Bruto/m²'] = (dados['Valor Bruto'] / dados['Area']).round(2)
dados.head(10)
```

```python
# Criando uma lista com os tipos que serão agrupados na categoria 'Casa'
casa = ['Casa', 'Casa de Condomínio', 'Casa de Vila']
```

```python
# Criando variável 'Tipo Agregado' com o método apply e por meio de uma função lambda: dados que estão na lista casa são
# classificados como 'Casa', os demais são 'Apartamento'
dados['Tipo Agregado'] = dados['Tipo'].apply(lambda x: 'Casa' if x in casa else 'Apartamento')
dados.head(25)
```

## Excluindo Variáveis

```python
# Criando um dataframe auxiliar com colunas específicas
dados_aux = pd.DataFrame(dados[['Tipo Agregado', 'Valor/m²', 'Valor Bruto', 'Valor Bruto/m²']])
dados_aux.head(10)
```

```python
# 1º método de exclusão de variáveis: del
del dados_aux['Valor Bruto']
dados_aux.head(10)
```

```python
# 1º método de exclusão de variáveis: pop
dados_aux.pop('Valor Bruto/m²')
dados_aux.head(10)
```

```python
# 3º método de exclusão de variáveis: drop. No caso do método drop(), precisaremos especificar o eixo em que se dará a 
# exclusão, uma vez que ele pode atuar especificamente em linhas ou colunas. Então, escreveremos axis = 1, afinal o eixo 0 
# corresponde às linhas e o 1 colunas. Precisaremos, ainda, assinalar inplace como True.
dados.drop(['Valor Bruto', 'Valor Bruto/m²'], axis = 1, inplace=True)
dados.head(10)
```

```python
# Exportando o novo banco de dados, sem o index
dados.to_csv('dados/aluguel_residencial_2.csv', sep=';', index=False)
```
