```python
import pandas as pd
```

```python
# Resultados dos lançamentos de moedas. c = cara e C = coroa
m1 = 'CCcCCccCCCccCcCccCcCcCCCcCCcccCCcCcCcCcccCCcCcccCc'
m2 = 'CCCCCccCccCcCCCCccCccccCccCccCCcCccCcCcCCcCccCccCc'
m3 = 'CccCCccCcCCCCCCCCCCcccCccCCCCCCccCCCcccCCCcCCcccCC'
m4 = 'cCCccCCccCCccCCccccCcCcCcCcCcCcCCCCccccCCCcCCcCCCC'
m5 = 'CCCcCcCcCcCCCcCCcCcCCccCcCCcccCccCCcCcCcCcCcccccCc'
```

```python
# Opção A: 
#
# eventos = {'m1': m1, 
#            'm2': m2, 
#            'm3': m3, 
#            'm4': m4, 
#            'm5': m5}
#moedas = pd.DataFrame(eventos)
#df = pd.DataFrame(data = ['Cara', 'Coroa'], 
#                    index = ['c', 'C'], 
#                    columns = ['Faces'])
#for item in moedas:
#    df = pd.concat([df, moedas[item].value_counts()], 
#                    axis = 1)
#df

# Criando um dicionário chamado eventos, com os resultados de cada moeda
eventos = {'m1': m1, 
           'm2': m2, 
           'm3': m3, 
           'm4': m4, 
           'm5': m5}
eventos
```

```python
# Gerando um dataframe a partir de eventos
moedas = pd.DataFrame(eventos)
moedas
```

### Aqui já encontramos o erro. É necessário passar uma lista para criar o dataframe, e não um dicionário.
## Alternativa A incorreta

```python
# Opção B:
#
# eventos = {'m1': list(m1), 
#            'm2': list(m2), 
#             'm3': list(m3), 
#             'm4': list(m4), 
#             'm5': list(m5)}
# moedas = pd.DataFrame(eventos)
# df = pd.DataFrame(data = ['Cara', 'Coroa'], 
#                     index = ['c', 'C'], 
#                     columns = ['Faces'])
# for item in moedas:
#     df = pd.concat([df, moedas[item].value_counts()], 
#                     axis = 1)
# df

# Criando os dados na variável eventos, mas dessa vez passando os valores dos testes como lista
eventos = {'m1': list(m1), 
            'm2': list(m2), 
            'm3': list(m3), 
            'm4': list(m4), 
            'm5': list(m5)}
eventos
```

```python
# Criando dataframe a partir de eventos
moedas = pd.DataFrame(eventos)
moedas
```

```python
# Criando um novo dataframe
df = pd.DataFrame(data = ['Cara', 'Coroa'], 
                  index = ['c', 'C'], 
                  columns = ['Faces'])
df
```

```python
# Concatenando os dataframes df com a frequencia de cada item no dataframe moedas (m1, m2, m3, m4, m5) com relação às colunas
# (axis = 1)
for item in moedas:
    df = pd.concat([df, moedas[item].value_counts()], 
                    axis = 1)
df
```

## Alternativa B correta

```python
# Opção C:
#
# eventos = {'m1': list(m1), 
#             'm2': list(m2), 
#             'm3': list(m3), 
#             'm4': list(m4), 
#             'm5': list(m5)}
# moedas = pd.DataFrame(eventos)
# df = pd.DataFrame(data = ['Cara', 'Coroa'], 
#                     index = ['c', 'C'], 
#                     columns = ['Faces'])
# for item in moedas:
#     df = pd.concat([df, moedas[item].sum()], 
#                     axis = 1)
# df

for item in moedas:
     df = pd.concat([df, moedas[item].sum()], 
                     axis = 1)
df
```

### Nesta opção não temos a frequência de cada lançamento, mas sim o somatório, o que retorna erro pois não temos dados numéricos
