## Mais obre Gráficos

```python
# Importando Matplotlib, Pandas, configurando matplotlib, importando banco de dados e verificando os dados importados
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
plt.rc('figure',figsize=(15,8))

dados = pd.read_csv('dados/aluguel.csv', sep = ';')
dados.head(10)
```

```python
# Criando área do gráfico e subdividindo o mesmo em 4 subgráficos.
area = plt.figure()
g1 = area.add_subplot(2, 2, 1)
g2 = area.add_subplot(2, 2, 2)
g3 = area.add_subplot(2, 2, 3)
g4 = area.add_subplot(2, 2, 4)

# O primeiro gráfico é um gráfico de dispersão (scatter plot) do valor dos imóveis pela área
g1.scatter(dados.Valor, dados.Area)
g1.set_title('Valor x Área')

# O segundo gráfico é um histograma dos valores
g2.hist(dados.Valor)
g2.set_title('Histograma')

# O terceiro gráfico apresenta 100 amostras aleatórias de 'Valor'. O index é renumerado para melhorar o gráfico.
dados_g3 = dados.Valor.sample(100) 
dados_g3.index = range(dados_g3.shape[0])
g3.plot(dados_g3)
g3.set_title('Amostra (Valor)')

# O quarto gráfico é um gráfico de barras da média dos valores por tipo
grupo = dados.groupby('Tipo')['Valor']
label = grupo.mean().index
valores = grupo.mean().values
g4.bar(label, valores)
g4.set_title('Valor Médio por Tipo')
```

```python
# Salvando figura. O argumento bbox_inches = 'tight' remove as bordas brancas em excesso da figura
area.savefig('dados/grafico.png', dpi = 300, bbox_inches = 'tight')
```
